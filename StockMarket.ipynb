```python
import pandas_datareader as pdr #抓取股票的api數據
import datetime #設定開始、結束日期
import matplotlib.pyplot as plt #2D圖形，視覺化開盤價、收盤價，plt自命名 
import numpy as np #股票市場分析的線性代數、統計(返回陣列)
%matplotlib inline 

start = datetime.datetime(2015,1,1)#開始日
end = datetime.datetime(2020,1,1)#結束日

#股票 = 從yahoo股市擷取('股票名稱','開始日','結束日')
twiter = pdr.get_data_yahoo('TWTR',start,end)
facebook = pdr.get_data_yahoo('FB',start,end)
instagram = pdr.get_data_yahoo('IG',start,end)

Amazon = pdr.get_data_yahoo('AMZN',start,end)
Walmart = pdr.get_data_yahoo('WMT',start,end)
Costco = pdr.get_data_yahoo('COST',start,end)
Kroger = pdr.get_data_yahoo('KR',start,end)
```

```python
#下載csv檔，方便jupyter查閱、處理數據
twiter.to_csv('TWTR_stock.csv')
facebook.to_csv('FB_stock.csv')
instagram.to_csv('FB_stock.csv')
```

```python
"""計算twite 5天,60天與240天的移動平均線"""
#擷取日線的收盤價數據，rolling(數據參數)
twiter['MA_5'] = twiter['Open'].rolling(5).mean()
twiter['MA_20'] = twiter['Open'].rolling(60).mean()
twiter['MA_240'] = twiter['Open'].rolling(240).mean()

twiter['Open'].plot(figsize=(15,7))#先畫收盤價的尺寸圖

#圖表說明的文字
twiter['MA_5'].plot(label='MA_5')
twiter['MA_20'].plot(label='MA_20')
twiter['MA_240'].plot(label='MA_240')

plt.legend() #圖表說明，預設英文，放置最前
plt.grid(True) #網格校正
plt.title('twiter MACD',fontsize = 16) #標題
plt.show()#秀圖表框，放置最後
```

```python
"""散佈矩陣圖 探討三個股間存在的依賴關係"""
from pandas.plotting import scatter_matrix
import pandas as pd

social_app = pd.DataFrame(np.random.randn(2000, 3)) #np.random.randn(密度, 比較項目)，columns 數據
social_app.columns=['TWTR Open','FB Open','IG Open'] 
pd.plotting.scatter_matrix(social_app, figsize=(8,8),alpha=0.2,hist_kwds={'bins':500}) #figsize 尺寸，alpha 透明度，hist_kwds直方圖{離散程度}
plt.show()
```
